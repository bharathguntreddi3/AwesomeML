```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 6, 11)
y = x**2
print(x)
print('\n')
print(y)
```

```python
plt.plot(x,y,'r') # red color
plt.xlabel('x axis')
plt.ylabel('y label')
plt.title('Title')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'b')
plt.subplot(1,2,2)    # 2 indicates new graph 1 represents in the first graph
plt.plot(y,x,'r')
```

```python
fig = plt.figure()   # create a figure/canvas
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])   # define add axes
axes.plot(x,y,'b')
axes.set_xlabel('x label')
axes.set_ylabel('y label')
axes.set_title('title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # small/inside axes

# main axes
axes1.plot(x, y, 'b')
axes1.set_xlabel('X_label')
axes1.set_ylabel('Y_label')
axes1.set_title('main axes')

# inside axes
axes2.plot(y, x, 'r')
axes2.set_xlabel('X_inside_label')
axes2.set_ylabel('Y_inside_label')
axes2.set_title('axes inside');
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
fig    
plt.tight_layout()
```

```python
# size, aspect ratio and dpi(dots per inch)
fig = plt.figure(figsize=(98,4), dpi=100)   # figsize defines the graph size
fig, axes = plt.subplots(figsize=(12,3))
axes.plot(x,y,'b')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, label='x**2')
ax.plot(x, x**3, label='x**3')
ax.legend(loc=0)    # decides optimal loation
# loc = 1    # upper right corner
# loc = 2    # upper left corner
# loc = 3    # lower left corner
# loc = 4    # lower right corner
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**2, 'r.-')   # red line with dots
ax.plot(x, x**3, 'b--')   # green dashed lines
ax.plot(x, x**4, color='#808080')    # color=parameter can also be used
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# possible linestype options ‘-‘, ‘–’, ‘-.’, ‘:’, ‘steps’
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```

```python
fig, axes = plt.subplots(figsize=(12, 4))

axes.plot(x, x**2, x, x**3)
axes.set_ylim([0, 60])
axes.set_xlim([2, 5])
axes.set_title("custom axes range");
```
