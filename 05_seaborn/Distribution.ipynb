```python
# Seaborn comes with some built-in datasets
import seaborn as sns
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python

### distplot
### shows the distribution of a universe set of observations
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=30)
```

```python
### jointplot
### allows to basically match up two distplots for bivariate data with your choicee
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
### pairplot
### plots pairwise relationships accross an entire dataframe
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

```python
### rugplot
### these plots are actually a very simple concept they just draw a clash marker for every point on a univariate distribution they are building blocks of KDE plots
sns.rugplot(tips['total_bill'])
```

```python
### KDE plot
### kde plots are 'Kernal Density Estimation plots' these KDE plots replace every single observation with a gaussian distribution centered around that value
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
# create dataset/frame
dataset = np.random.randn(25)
# create a rugplot
sns.rugplot(dataset)
# set x-axis for the plot
x_min = dataset.min() - 2
x_max = dataset.max() + 2
# 100 equally spaced points from x_min to x_max
x_axis = np.linspace(x_min, x_max, 100)
# set bandwidth info on this-
url = 'http://en.wikipedia.org/wiki/Kernel_density_estimation#Practical_estimation_of_the_bandwidth'
bandwith = ((4*dataset.std()**5)/(3*len(dataset)))**.2
# create an empty kernel list
kernel_list = []
# plot each basis function
for data_point in dataset:
    # create a kernel for each point and append to the list
    kernel = stats.norm(data_point, bandwith).pdf(x_axis)
    kernel_list.append(kernel)
    # scale for plotting
    kernel = kernel/kernel.max()
    kernel = kernel*.4
    plt.plot(x_axis, kernel, color = 'grey', alpha=0.5)
```

```python
# to pot the kde plot we can sum these basis functions
# pot the sum of the basis function 
sum_of_kde = np.sum(kernel_list, axis=0)
# plot figure 
fig = plt.plot(x_axis, sum_of_kde, color = 'indianred')
# add the initial rugplot
sns.rugplot(dataset, c = 'indianred')
# get rid of y-tick marks
plt.yticks([])
# set title 
plt.suptitle("SUm of the basis Functions")
```
