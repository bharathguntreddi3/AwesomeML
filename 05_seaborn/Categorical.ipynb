```python
### Categorical plot
# factorplot
# boxplot
# violinplot
# stripplot
# swarmplot
# barplot
# countplot
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
### barplot
### aggregate the categorcial data based off some function
sns.barplot(x='sex', y= 'total_bill', data = tips)
```

```python
import numpy as np
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
### countplot
### estimates the same as barplot except the estimator is explicitly counting the number of occurences
sns.countplot(x='sex', data = tips)
```

```python
### boxplot 
### used to show distributions of the categorical data
sns.boxplot(x='day', y='total_bill', data=tips, palette='rainbow')
```

```python
# with orient h- hroizontal
sns.boxplot(data = tips, palette='rainbow', orient='h')
```

```python
sns.boxenplot(x='day', y='total_bill', hue='smoker', data=tips, palette='coolwarm')
```

```python
### violinplot
### shows the distribution of the quantitative data across several levels of one categorical variables such that those distributions can be compared
sns.violinplot(x='day', y='total_bill', data = tips, palette='rainbow')
```

```python
sns.violinplot(x='day', y='total_bill', data = tips, hue='sex', palette='Set1')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True, palette='Set1')
```

```python
### striplot
### draws a sscatter plot where one variable is vatehorical
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', palette='Set1')
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex', palette='Set1')
```

```python
### swarmplot
### similar to stripplot but the points are adjusted so that they don't overlap
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sns.swarmplot(x='day', y='total_bill', hue='sex', data=tips, palette='Set1')
```

```python
### combining categorical plots
sns.violinplot(x='tip', y='day', data=tips, palette='rainbow')
sns.swarmplot(x='tip', y='day', data=tips, color='black', size=3)
```
