## Matrixs

```python
### MATRIX PLOTS
### matrix plots allows to plot data as color-encoded matrices and can indicate clusters within data
```

```python
import seaborn as sns
```

```python
flights = sns.load_dataset('flights')
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
flights.head()
```

```python
### heatmap
### your data should be in matrix form heatmap colors them
tips.corr()   # correlation data
```

```python
sns.heatmap(tips.corr())
```

```python
sns.heatmap(tips.corr(), cmap='coolwarm', annot=True)
```

```python
pvflights = flights.pivot_table(values='passengers', index='month', columns='year')
```

```python
sns.heatmap(pvflights)
```

```python
sns.heatmap(pvflights, cmap='magma', linecolor='white', linewidths=3)
```

```python
### cluster map
### uses heirarchal clustering to produce a clustered version of heatmap
sns.clustermap(pvflights)
```

```python
# normalization
sns.clustermap(pvflights, cmap='coolwarm', standard_scale=1)
```

## GRIDS

```python
### grids
### helps to create similar plots separated by features
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
### pairgrid
### plot pairwise relationships
sns.PairGrid(iris)
```

```python
### map to the grid
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
### map to lower upper and diagonal
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(plt.hist)
```
