```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.set_style('white')  #ticks
sns.countplot(x='sex',data=tips)
```

```python
sns.despine()   # removes the spine
```

```python
sns.countplot(x='sex',data=tips)
sns.despine(left=True)
# removes the spine top or left 
```

```python
# Non Grid Plot
plt.figure(figsize=(12,3))          #changes the figure size
sns.countplot(x='sex',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',aspect=4,data=tips)
```

```python
sns.set_context('poster',font_scale=4)
sns.countplot(x='sex',data=tips,palette='coolwarm')
# can increase the font size also
```
